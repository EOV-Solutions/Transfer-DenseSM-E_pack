```python
import pandas as pd 
import numpy as np 
import os
```

```python
def normalizingData(X, min_per, max_per):
    temp=(X - min_per) / (max_per - min_per)
    temp[temp>1]=1
    temp[temp<0]=0
    return temp
```

```python
folder = '1km_global_data/india/ndvi/ndvi_output'
output_folder = '1km_global_data/india/ndvi/ndvi_normalized'
os.makedirs(output_folder, exist_ok=True)
for file in os.listdir(folder):
    print(file)
    file_path = os.path.join(folder, file)
    output_path = os.path.join(output_folder, file)

    df_ndvi = pd.read_csv(file_path)
    ndvi_normalized = normalizingData(np.asarray(df_ndvi['NDVI']), -0.2, 1.0)
    df_ndvi['NDVI'] = ndvi_normalized 

    # renamed 'Unamed: 0' column to 'date' 
    if 'Unnamed: 0' in df_ndvi.columns:
        df_ndvi = df_ndvi.rename(columns={'Unnamed: 0': 'date'})
    df_ndvi.to_csv(output_path)
```

```python
sm_folder = '1km_global_data/india/output_india_1km'
ndvi_normalized_folder = '1km_global_data/india/ndvi/ndvi_normalized'
output_folder = '1km_global_data/india/ndvi/output_india_landsat'
os.makedirs(output_folder, exist_ok = True)

for file in os.listdir(sm_folder):
    sm_file_path = os.path.join(sm_folder, file)
    ndvi_normalized_path = os.path.join(ndvi_normalized_folder, file)

    df_sm = pd.read_csv(sm_file_path)
    df_ndvi_normalized = pd.read_csv(ndvi_normalized_path)

    df_ndvi_normalized = df_ndvi_normalized.sort_values('date').reset_index(drop = True) 

    for date in df_sm['date']:
        last_index =  df_ndvi_normalized[df_ndvi_normalized['date'] <= date].last_valid_index()

        if last_index is None or last_index < 45:
            continue 
        ndvi_values = df_ndvi_normalized.loc[(last_index - 45):last_index, 'NDVI'].values 
        idx = df_sm.index[df_sm['date'] == date] 
        if len(idx) > 0:
            df_sm.loc[idx[0], '15':'60'] = ndvi_values 

    # Export updated output file
    output_path = os.path.join(output_folder, file)
    df_sm.to_csv(output_path, index = False)
```
